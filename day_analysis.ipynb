```python
import pandas as pd

with open('Project.csv', 'r') as f:
    df = pd.read_csv(f)
selected_features_df = df[['Day of Week', 'Total_Bill']]
selected_features_df
```

```python
# print(selected_features_df['Day of Week'].unique())
group_by = selected_features_df.groupby('Day of Week')['Total_Bill'].sum()
group_by
```
